<h1 style="font-family: 'Times New Roman'; letter-spacing: 0.08em;">Importing the Libraries</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.linear_model
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

<h1 style="font-family: 'Times New Roman'; letter-spacing: 0.08em;">Loading the Data</h1>

```python
pts = np.loadtxt('linpts.txt')
X = pts[:,:2]
Y = pts[:,2].astype('int')
```

<h1 style="font-family: 'Times New Roman'; letter-spacing: 0.08em;">Machine Learning Models</h1>

<h2 style="font-family: 'Times New Roman'; letter-spacing: 0.05em;">Logistic Regression</h2>

```python
LR = sklearn.linear_model.LogisticRegression().fit(X, Y)

plt.rcParams.update(plt.rcParamsDefault)
_, ax = plt.subplots(figsize=(8, 6))
DecisionBoundaryDisplay.from_estimator(
    LR,
    X,
    cmap=plt.cm.OrRd,
    ax=ax,
    response_method="predict",
    plot_method="pcolormesh",
    shading="auto",
    eps=0.5,
)
plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolors="k", cmap=plt.cm.OrRd)
```

<h2 style="font-family: 'Times New Roman'; letter-spacing: 0.05em;">Support Vector Classifier</h2>

```python
svc = SVC(kernel='linear').fit(X, Y)
_, ax = plt.subplots(figsize=(8, 6))
DecisionBoundaryDisplay.from_estimator(
    svc,
    X,
    cmap=plt.cm.YlGn,
    ax=ax,
    response_method="predict",
    plot_method="pcolormesh",
    shading="auto",
    eps=0.5,
)
plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolors="k", cmap=plt.cm.YlGn)
```

<h2 style="font-family: 'Times New Roman'; letter-spacing: 0.05em;"> Decision Tree</h2>

```python
dst = DecisionTreeClassifier().fit(X,Y)
_, ax = plt.subplots(figsize=(8, 6))
DecisionBoundaryDisplay.from_estimator(
    dst,
    X,
    cmap=plt.cm.RdPu,
    ax=ax,
    response_method="predict",
    plot_method="pcolormesh",
    shading="auto",
    eps=0.5,
)
plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolors="k", cmap=plt.cm.RdPu)
```

<h2 style="font-family: 'Times New Roman'; letter-spacing: 0.05em;">Naive Bayes</h2>

```python
gnb = GaussianNB().fit(X,Y)
_, ax = plt.subplots(figsize=(8, 6))
DecisionBoundaryDisplay.from_estimator(
    gnb,
    X,
    cmap=plt.cm.PuBu,
    ax=ax,
    response_method="predict",
    plot_method="pcolormesh",
    shading="auto",
    eps=0.5,
)
plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolors="k", cmap=plt.cm.PuBu)
```

<h2 style="font-family: 'Times New Roman'; letter-spacing: 0.05em;">KNeighbors</h2>

```python
knn = KNeighborsClassifier(n_neighbors=3).fit(X,Y)
_, ax = plt.subplots(figsize=(8, 6))
DecisionBoundaryDisplay.from_estimator(
    knn,
    X,
    cmap=plt.cm.GnBu,
    ax=ax,
    response_method="predict",
    plot_method="pcolormesh",
    shading="auto",
    eps=0.5,
)
plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolors="k", cmap=plt.cm.GnBu)
```
